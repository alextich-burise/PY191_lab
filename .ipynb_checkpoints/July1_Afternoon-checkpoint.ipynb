#### Define a circle function that will return true if in the circle and false otherwise
#### For this simulation, we will only be working in the first quartile and then multiplying by 4 at the end

```python
def circle(x,y):
    return x**2 + y**2 <= 1
```

```python
import random
iterations = 100_000_000
number = 0
for i in range(iterations):
    x = random.random()
    y = random.random()
    if circle(x,y):
        number+=1
print(4 * number/iterations)
```

```python
text = "all happy families are alike each unhappy family is unhappy in its own way".replace(" ", "")

letters = {chr(ord("a")+x):0 for x in range(26) }
for letter in letters.keys():
    for char in text:
        letters[letter]+= 1 if char==letter else 0
letters
```

```python
para = "It was the best of times, it was the worst of times... in short, the period was so far like the present period, that some of its noisiest authorities insisted on its being recieved, for good or for evil, in the superlative degree of comparison only"
para = para.replace(",", "")
para = para.replace(".", "")
para = para.split(" ")
para = sorted(para)
words = {word:0 for word in para}
for key in words.keys():
    for word in para:
        words[key]+=1 if word == key else 0
words
```

```python
paragraph = "It was the best of times, it was the worst of times... in short, the period was so far like the present period, that some of its noisiest authorities insisted on its being recieved, for good or for evil, in the superlative degree of comparison only"
para = paragraph.replace(",", "")
para = para.replace(".", "")
para = para.split(" ")
para2 = [word[::((i+1)%2)*2-1] for i, word in enumerate(para)]
paragraph = paragraph.split(" ")
for idx, p in enumerate(paragraph):
    paragraph[idx] = paragraph[idx].replace(para[idx], para2[idx])
paragraph = " ".join(paragraph)
paragraph
```
